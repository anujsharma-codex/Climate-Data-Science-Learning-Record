# Figure Size and DPI and Aspect Ratio

```python
import matplotlib.pyplot as plt

# Wide figure (good for time series)
fig, ax= plt.subplots(figsize=(12,4))
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_title('Wide Figute - Good for Long Time Series')

# Square figure (good for heatmaps, scatter plots)
fig, ax= plt.subplots(figsize=(6,6))
ax.scatter([1,2,3,4],[1,4,2,3])
ax.set_title('Square Figure - Good for Scatter Plots')
```

# Adding legend

```python
# Legend tells which line reps which graph

import numpy as np
x=np.linspace(0,5,11)
y=x**2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend()

# for specifying location
```

```python
# (x,y) coordinates where 0=left/bottom, 1=right/top
# Bottom right corner
ax.legend(bbox_to_anchor=(1,0))

# Top right corner
ax.legend(bbox_to_anchor= (1,1))

# Outside the plot (right side)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Bottom center outside
ax.legend(bbox_to_anchor=(0.5, -0.1), loc='upper center')
```

```python
# Legend with Title and Customization

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, np.sin(x), label='sin(x)', color='blue')
ax.plot(x, np.cos(x), label='cos(x)', color='red')
ax.plot(x, np.tan(x), label='tan(x)', color='green')

ax.legend(
    loc='upper left',
    title='Trig Functions',
    fontsize=10,
    title_fontsize=12,
    frameon=True,
    shadow=True,
    fancybox=True
)
plt.ylim(-2, 2)
plt.show()
```

# Setting Colours with Matplotlib

```python
import matplotlib.pyplot as plt

fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color= 'darkred', lw=1, alpha = 1, linestyle= '-', marker='o', markersize= 20,
       markerfacecolor='black', markeredgewidth=3, markeredgecolor='red')
# Linewidth or lw thickens the graph plot
# color cn be written in RGB Hex codes as well.
# alpha= controls transparency (0-1)
# linestyle or ls controls the style of the line.
# marker is used to mark the points on the plot
# markersize decided the largeness or smallness of the marker.
# markerfacecolor decides the color of the marker.
# markeredgewidth decided the width of border of the markers
# markeredgecolor decided the color of the boundary of marker

# Now suppose we want to show the plot only between x=1 and x=2
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
