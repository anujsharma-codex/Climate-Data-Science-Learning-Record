```python
import numpy as np
arr1=np.arange(0,20,dtype=np.float16).reshape(4,5)
print(f"arr1: \n{arr1}")
print(f"dtype: {arr1.dtype}")
arr2=arr1.astype(np.float32)
print(f"dtype now: {arr1.dtype}")
print("-----------------------------------")

arr1_recovered=arr2.astype(np.float16)
loss = arr1_recovered-arr1
print(f"arr1: \n{arr1}")
print(f"arr2: \n{arr2}")
print(f"arr1_recovered: \n{arr1_recovered}")
print(f"loss : \n{loss}")
print(f"cumulatve precision loss: {np.sum(loss)}")
```

```python
import numpy as np
arr1=np.arange(1,7).reshape(1,6)
col=np.arange(31,37).reshape(6,1)
table=arr1*col
print(f"table: \n{table}")
print()

print("# checking for overflow...")
print()

table_twice=table*2
print(f"table x 2: \n{table_twice}")
overflow_state=np.any(table_twice>255)
print()

if overflow_state:
    print(f"overflow detected at : \n{np.argwhere(table_twice>255)}")
else:
    print("no overflow")
```
