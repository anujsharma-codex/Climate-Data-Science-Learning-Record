# Standard Deviation

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({
    "Name": ["Mohan","Andrea","Sofia","Joe","Virat","Abdul"],
    "Score": [75, 72, 68, 65, 67, 73]
})

avg= df.Score.mean()
print(avg)
```

```python
# This happens by simply subtracting the data points from the datapoints

df["Abs"]=(df.Score - avg).abs()
print(df)
```

```python
std_dev = df["Abs"].mean()
print(std_dev)
```

```python
# one more example
df2=pd.DataFrame({
    "Name":["Mohan","Andrea","Sofia","Joe","Virat","Abdul"],
    "Score":[93,96,43,47,51,90],
    "Abs":[23,26,27,23,19,20]
})

print(f"Standard Deviation Here: {df2.Abs.mean()}")
```
