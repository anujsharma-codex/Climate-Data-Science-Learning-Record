# groupby()

```python
# e.g.
import pandas as pd

# Sample data
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C'],
    'Value': [10, 20, 30, 40, 50]
})

print(df)
```

```python
The code :
result = df.groupby('Category')['Value'].mean()
print(result)
```

```python
# All Aggregation functions-
# Basic statistics
df.groupby('Category')['Value'].mean()   # Average
df.groupby('Category')['Value'].sum()    # Total
df.groupby('Category')['Value'].count()  # Number of rows
df.groupby('Category')['Value'].min()    # Minimum
df.groupby('Category')['Value'].max()    # Maximum
df.groupby('Category')['Value'].std()    # Standard deviation
df.groupby('Category')['Value'].var()    # Variance
df.groupby('Category')['Value'].median() # Middle value

# First and last
df.groupby('Category')['Value'].first()  # First value in group
df.groupby('Category')['Value'].last()   # Last value in group

# Unique
df.groupby('Category')['Value'].nunique() # Number of unique values
```

```python
# e.g.s

import pandas as pd
import numpy as np

# 1. Sales Data
sales = pd.DataFrame({
    'Product': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [100, 200, 150, 250, 300, 180]
})

# Total sales by product
sales.groupby('Product')['Sales'].sum()
# Product
# A    600
# B    580

# Average sales by region
sales.groupby('Region')['Sales'].mean()
# Region
# North    183.33
# South    210.00

# Sales by both product and region
sales.groupby(['Product', 'Region'])['Sales'].sum()
# Product  Region
# A        North    400
#          South    200
# B        North    150
#          South    430

# ------------------------------------------------------------------

# 2. Weather Data

weather = pd.DataFrame({
    'City': ['NYC', 'NYC', 'LA', 'LA', 'NYC', 'LA'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Mar', 'Mar'],
    'Temp': [32, 35, 65, 68, 45, 70]
})

# Average temperature by city
weather.groupby('City')['Temp'].mean()
# City
# LA    67.67
# NYC   37.33

# Min temperature by month
weather.groupby('Month')['Temp'].min()
# Month
# Feb    35
# Jan    32
# Mar    45
```
