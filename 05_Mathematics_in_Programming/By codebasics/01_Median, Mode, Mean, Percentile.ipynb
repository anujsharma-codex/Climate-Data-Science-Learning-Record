# Basic use of Mean and Median in simple descriptive statistics

```python
import pandas as pd
import numpy as np

incomes= pd.DataFrame({
        "Name":["Rob", "Rafiq", "Nina", "Sofia", "Mohan", "Tao"],
    "Monthly Income $":[5000,6000,4000,7500,8000,7000]
})
print(incomes)
print()
print(f"Mean : {np.mean(incomes["Monthly Income $"])}")
```

```python
incomes= pd.DataFrame({
        "Name":["Rob", "Rafiq", "Nina", "Sofia", "Mohan", "Tao","Elon Musk"],
    "Monthly Income $":[5000,6000,4000,7500,8000,7000, 10000000]
})
print(incomes)
print()
print(f"Mean : {np.mean(incomes["Monthly Income $"])}")
```

```python
incomes.sort_values(by="Monthly Income $", inplace=True, ascending=True)
print(incomes)
print()
print(f"Median: {incomes["Monthly Income $"].median()}")
```

# Use case 2:
# Handling Missing Values (Data Cleaning (Filling NA Values))

```python
df= pd.DataFrame({
    "Name":["Rob", "Rafiq", "Nina", "Sofia", "Mohan", "Tao","Elon Musk"],
    "Monthly Income $":[5000,6000,4000, None ,8000,7000, 10000000],
    "Credit Score":[650, 400, 780, 810, 410, 850, 880],
    "Approve Loan":["No","No","Yes","Yes","No","Yes","Yes"]
})
print(df)
```

```python
df["Monthly Income $"] = df["Monthly Income $"].fillna(df["Monthly Income $"].median())
print(df["Monthly Income $"])
```

# Removing Outliers

```python
# Calculation

df_99 = df["Monthly Income $"].quantile(0.99)
clean_df = df[df["Monthly Income $"] <= df_99]
print(clean_df)  # cleaned df without any outlier
print()

# Use of quantile function for getting the percentile
print("25th percentile in the incomes column is :-")
print(df["Monthly Income $"].quantile(0.25))
```

```python
# we can get these also by using .describe() function:

clean_df.describe()
```
