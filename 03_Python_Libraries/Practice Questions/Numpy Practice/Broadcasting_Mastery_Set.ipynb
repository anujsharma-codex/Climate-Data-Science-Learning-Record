# BROADCASTING - Complete Mastery Set

## The Rules You Must Internalize
Before starting, memorize these 3 rules:

1. Arrays are compatible if dimensions match OR one is 1
2. Dimensions are compared from RIGHT to LEFT
3. Missing dimensions are treated as 1

**Q6. Create a 3×4 array of zeros. Add the scalar 5 to it. What shape is the result? Why does this work?**

````python
import numpy as np
arr_zero=np.zeros((3,4)) + 5
print(arr_zero)
# this works because a scalar is treated as a 1D-array of one element. and broadcasting works with all scalars
````

### The newaxis Trick

**Q11. You have `a.shape = (10,)` and `b.shape = (10, 10)`. You want to add a to each row of b. How do you reshape a using `np.newaxis`?**

````python
# for reshaping

a=np.random.randint(1,11,size=10)
b=np.random.randint(1,50,size=100).reshape(10,10)

# incompatible addinig:
# res=a+b

# compatible adding using np.newaxis
a[:,np.newaxis]

# or
# a[:,None]

result=a+b
print(result)
````

**Q14. Convert a 1D array of shape (12,) to shapes: (12,1), (1,12), (12,1,1), (1,12,1), (1,1,12) using `np.newaxis`.**

````python
arr=np.random.randint(1,13,size=12)

arr1=arr[:,np.newaxis]
print(arr1.shape)

arr2=arr[np.newaxis,:]
print(arr2.shape)

arr3=arr[:,np.newaxis,None]
print(arr3.shape)

arr4=arr[np.newaxis,:,np.newaxis]
print(arr4.shape)

arr5=arr[None,None,:]
print(arr5.shape)
````

**Q15. You have latitude (100,) and longitude (120,) arrays. Create a 2D grid of shape (100, 120) where each element is (lat, lon) using broadcasting and `np.newaxis`.**

````python
lat=np.random.randint(1,101,size=100)
lon=np.random.randint(1,121, size=120)
print(lat.shape,lon.shape)

dtype=[('lat',float),('lon',float)]
grid=np.empty((100,120),dtype=dtype)
grid['lat']=lat[:,np.newaxis]
grid['lon']=lon[np.newaxis,:]
print(grid)
````
